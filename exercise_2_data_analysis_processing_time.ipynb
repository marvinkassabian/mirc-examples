## Exercise #2: Data Analysis and Visualization

```python
import pandas as pd

# Load the data
exercise_2_df = pd.read_excel('data/Exercise #2.xlsx')

# Display the first few rows
exercise_2_df.head()
```

```python
# Attempt to convert 'event_date' to datetime, errors='coerce' will set invalid parsing to NaT
exercise_2_df['event_date'] = pd.to_datetime(exercise_2_df['event_date'], errors='coerce')

# Calculate time taken for each application
exercise_2_df['ProcessingTime'] = (
            exercise_2_df['application/petition outcome date (if applicable)'] - exercise_2_df['event_date']).dt.days

exercise_2_df
```

```python
exercise_2_df['ProcessingTime'].dropna().sort_values()
```

```python
exercise_2_df.dtypes
```

```python
# Display statistics about processing time
processing_time_stats = exercise_2_df['ProcessingTime'].describe()
processing_time_stats
```

```python
import matplotlib.pyplot as plt

# Create a histogram of processing times
plt.hist(exercise_2_df['ProcessingTime'].dropna(), bins=20, color='skyblue')
plt.title('Distribution of Processing Times')
plt.xlabel('Processing Time (days)')
plt.ylabel('Number of Applications')
plt.show()
```
